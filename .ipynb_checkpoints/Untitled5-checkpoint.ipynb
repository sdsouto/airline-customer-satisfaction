```python
import pandas as pd
import seaborn as sns

df1 = pd.read_csv('data/2017_Yellow_Taxi_Trip_Data.csv')
```

```python
df1
```

```python
# Create a scatter plot of duration and trip_distance, with a line of best fit
sns.set(style='whitegrid')
f = plt.figure()
f.set_figwidth(5)
f.set_figheight(5)
sns.regplot(x=df1['trip_distance'],
            y=df1['fare_amount'],
            scatter_kws={'alpha':0.5, 's':5},
            line_kws={'color':'red'})
plt.ylim(0, 70)
plt.xlim(0, 70)
plt.title('Scatterplot of mean distance and fare amount')
plt.show()
```
